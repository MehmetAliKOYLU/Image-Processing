```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# SORU 1-3'ün cevabı:

image = cv2.imread("aslan.jpg")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)


fig = plt.figure(figsize=(10, 7))

rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(image) 
plt.axis('off') 
plt.title("BGR Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(image_rgb) 
plt.axis('off') 
plt.title("RGB Image") 

fig.add_subplot(rows, columns, 3)
plt.imshow(image_gray,cmap="gray") 
plt.axis('off') 
plt.title("GRAY Image") 
```

```python
# 4-5. soruların fonksiyon olarak cevabı:


height, width = image_rgb.shape[:2]

def img_resize(height, width):
    if height > 600 and width > 600:
        print("Resmin yüksekliği ya da genişliği 600'den büyük olduğu için sadece resim boyutu küçültüldü.")
        new_image = cv2.resize(image_rgb, (width // 2, height // 2), interpolation=cv2.INTER_LANCZOS4)
        return new_image, None  # İkinci öğe None olarak döndürülüyor
    elif height < 600 and width < 600:
        print("Resmin yüksekliği ya da genişliği 600'den küçük olduğu için resim boyutu küçültüldü ve 2 katına çıkartıldı.")
        new_image = cv2.resize(image_rgb, (width // 2, height // 2), interpolation=cv2.INTER_LANCZOS4)
        new_image2 = cv2.resize(image_rgb, (width * 2, height * 2), interpolation=cv2.INTER_LANCZOS4)
        return new_image, new_image2
    else:
        new_img = cv2.resize(image_rgb, (width // 2, height // 2), interpolation=cv2.INTER_LANCZOS4)
        return new_img, None  # İkinci öğe None olarak döndürülüyor
```

```python
image_rgb.shape
```

```python

resized_img, resized_img2 = img_resize( height, width)

fig = plt.figure(figsize=(10, 7))

rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(image_rgb) 
plt.axis('off') 
plt.title("original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(resized_img) 
plt.title("small Image") 
# İkinci resim varsa boyutlarını yazdırın
if resized_img2 is not None:
    fig.add_subplot(rows, columns, 3)
    plt.imshow(resized_img2) 
    plt.title("big Image") 

```

```python
# İlk resim boyutlarını yazdırın
if resized_img is not None:
    print(f"Resized Görüntünün Yüksekliği = {resized_img.shape[0]}")
    print(f"Resized Görüntünün Genişliği = {resized_img.shape[1]}")

# İkinci resim varsa boyutlarını yazdırın
if resized_img2 is not None:
    print(f"İkinci Resized Görüntünün Yüksekliği = {resized_img2.shape[0]}")
    print(f"İkinci Resized Görüntünün Genişliği = {resized_img2.shape[1]}")
```

```python
img_flag_russia = np.zeros((300,420,3),dtype=np.uint8)

# ilk 100 satırın siyah görünmesini sağlayan kod:
# [Red, Green, Blue]
img_flag_russia[0:100,:] = [220,220,220]

# 100. satırdan 200. satıra kadar kırmızı görünmesini sağlayan kod:
img_flag_russia[100:200,:] = [0,0,255]

# 200. satırdan 300. satıra kadar (son satıra kadar)
# sarı görünmesini sağlayan kod:
img_flag_russia[200:,:] = [255,0,0]

plt.imshow(img_flag_russia)
plt.axis('off')
plt.title("RUSSIA FLAG")
```
