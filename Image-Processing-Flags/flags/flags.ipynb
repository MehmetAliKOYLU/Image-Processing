```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# SORU 1-3'n cevab:

image = cv2.imread("aslan.jpg")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)


fig = plt.figure(figsize=(10, 7))

rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(image) 
plt.axis('off') 
plt.title("BGR Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(image_rgb) 
plt.axis('off') 
plt.title("RGB Image") 

fig.add_subplot(rows, columns, 3)
plt.imshow(image_gray,cmap="gray") 
plt.axis('off') 
plt.title("GRAY Image") 
```

```python
# 4-5. sorularn fonksiyon olarak cevab:


height, width = image_rgb.shape[:2]

def img_resize(height, width):
    if height > 600 and width > 600:
        print("Resmin ykseklii ya da genilii 600'den byk olduu iin sadece resim boyutu kltld.")
        new_image = cv2.resize(image_rgb, (width // 2, height // 2), interpolation=cv2.INTER_LANCZOS4)
        return new_image, None  # kinci e None olarak dndrlyor
    elif height < 600 and width < 600:
        print("Resmin ykseklii ya da genilii 600'den kk olduu iin resim boyutu kltld ve 2 katna kartld.")
        new_image = cv2.resize(image_rgb, (width // 2, height // 2), interpolation=cv2.INTER_LANCZOS4)
        new_image2 = cv2.resize(image_rgb, (width * 2, height * 2), interpolation=cv2.INTER_LANCZOS4)
        return new_image, new_image2
    else:
        new_img = cv2.resize(image_rgb, (width // 2, height // 2), interpolation=cv2.INTER_LANCZOS4)
        return new_img, None  # kinci e None olarak dndrlyor
```

```python
image_rgb.shape
```

```python

resized_img, resized_img2 = img_resize( height, width)

fig = plt.figure(figsize=(10, 7))

rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(image_rgb) 
plt.axis('off') 
plt.title("original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(resized_img) 
plt.title("small Image") 
# kinci resim varsa boyutlarn yazdrn
if resized_img2 is not None:
    fig.add_subplot(rows, columns, 3)
    plt.imshow(resized_img2) 
    plt.title("big Image") 

```
