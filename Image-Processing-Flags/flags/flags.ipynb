```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
# SORU 1-3'ün cevabı:

img_bgr = cv2.imread("./aslan.jpg")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

fig = plt.figure(figsize=(10, 7))

rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(img_bgr) 
plt.axis('off') 
plt.title("BGR Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(img_rgb) 
plt.axis('off') 
plt.title("RGB Image") 

fig.add_subplot(rows, columns, 3)
plt.imshow(img_gray,cmap="gray") 
plt.axis('off') 
plt.title("GRAY Image") 
```

```python
# 4-5. soruların fonksiyon olarak cevabı:

def my_img_resize(image:np.ndarray):
    height, width = image.shape[:2]

    if height>600 and width>600:
        new_img = cv2.resize(image,(width//2,height//2),cv2.INTER_LANCZOS4)
        return new_img

    elif height<600 and width<600:
        new_img_1 = cv2.resize(image,(width//2,height//2),cv2.INTER_LANCZOS4)
        new_img_2 = cv2.resize(image,(width*2,height*2),cv2.INTER_LANCZOS4)

        return new_img_1, new_img_2

    else:
        new_img = cv2.resize(image,(width//2,height//2),cv2.INTER_LANCZOS4)
        return new_img
```

```python
img_resized = my_img_resize(img_rgb)

plt.imshow(img_resized)
```

```python
#4 ve 5. soruların cevabı (Düz Kod olarak):

yukseklik, genislik = img_rgb.shape[:2]

if yukseklik>600 and genislik>600:
    my_resized_small = cv2.resize(img_rgb,(genislik//2,yukseklik//2),
                                 cv2.INTER_LANCZOS4)
    print("Her iki boyut da 600'den büyük")
    print("Sadece küçültme işlemi yapıldı. Değişkenler = my_resized_small")

elif yukseklik<600 and genislik<600:
    my_resized_small = cv2.resize(img_rgb,(genislik//2,yukseklik//2),
                                 cv2.INTER_LANCZOS4)

    my_resized_big = cv2.resize(img_rgb,(genislik*2,yukseklik*2),
                                 cv2.INTER_LANCZOS4)
    print("Her iki işlem de yapıldı. Değişkenler:")
    print("my_resized_small ve my_resized_big")

else:
    my_resized_small = cv2.resize(img_rgb,(genislik//2,yukseklik//2),
                                 cv2.INTER_LANCZOS4)
    print("Boyutlardan yalnızca biri 600'den büyük")
    print("Sadece küçültme işlemi yapıldı. Değişkenler = my_resized_small")
```

```python
# 300x420 boyutunda 3 kanallı içi sıfır ile dolu bir matris tanımlıyoruz:

img_flag_germany = np.zeros((300,420,3),dtype=np.uint8)

# ilk 100 satırın siyah görünmesini sağlayan kod:
# [Red, Green, Blue]
img_flag_germany[0:100,:] = [0,0,0]

# 100. satırdan 200. satıra kadar kırmızı görünmesini sağlayan kod:
img_flag_germany[100:200,:] = [255,0,0]

# 200. satırdan 300. satıra kadar (son satıra kadar)
# sarı görünmesini sağlayan kod:
img_flag_germany[200:,:] = [255,255,0]

plt.imshow(img_flag_germany)
plt.axis('off')
plt.title("GERMANY FLAG")
```
